```python
clean_df = df1.copy()
```

```python
clean_df["rate"] = clean_df["rate"].replace(["New", "-"],np.nan)
clean_df["rate"] = clean_df["rate"].replace(["NEW", "-"],np.nan)
```

```python
clean_df["rate"].unique()
```

```python
clean_df["rate"]= clean_df["rate"].str.replace("/5","" , regex =False)
```

```python
clean_df["rate"].unique()
```

```python
clean_df["rate"].head(10)
clean_df["rate"].tail(10)
```

```python
clean_df["rate"]= clean_df["rate"].astype(float)
```

```python
clean_df["rate"].dtype
```

```python
clean_df["approx_cost(for two people)"].head()
```

```python
clean_df["approx_cost(for two people)"].tail()
```

```python
clean_df["rate"].isnull().sum()
```

```python
clean_df["approx_cost(for two people)"] = clean_df["approx_cost(for two people)"].replace([",", ""],np.nan)
```

```python
clean_df["approx_cost(for two people)"].tail()
```

```python
clean_df["approx_cost(for two people)"] = pd.to_numeric(
    clean_df["approx_cost(for two people)"],
    errors="coerce"
)
```

```python
clean_df["approx_cost(for two people)"].tail()
```

```python
clean_df["approx_cost(for two people)"] = clean_df["approx_cost(for two people)"].astype("Int64")
```

```python
clean_df["approx_cost(for two people)"].dtype
```

```python
clean_df.isnull().sum()
```

```python
missing = (clean_df.isnull().sum() /len(clean_df))*100
missing.sort_values(ascending = False)
```

```python
clean_df["rate"] = clean_df["rate"].fillna(clean_df["rate"].median())
clean_df["approx_cost(for two people)"] = clean_df["approx_cost(for two people)"].fillna(clean_df["approx_cost(for two people)"].median())
clean_df["rest_type"] = clean_df["rest_type"].fillna(clean_df["rest_type"].mode()[0])
clean_df["cuisines"] = clean_df["cuisines"].fillna(clean_df["cuisines"].mode()[0])
clean_df["location"] = clean_df["location"].fillna(clean_df["location"].mode()[0])
```

```python
missing = (clean_df.isnull().sum() /len(clean_df))*100
missing.sort_values(ascending = False)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (8,5))
plt.hist(clean_df["rate"],bins = 10)
plt.title("Distribution of Resturant Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Resturant")
plt.show()
```

```python
top_location = clean_df["location"].value_counts().head(10)
plt.figure(figsize = (8,5))
top_location.plot(kind ="bar")
plt.xlabel("Location")
plt.title("Top 10 Locations by Number of Restaurants")

plt.ylabel("Restaurant Count")
plt.xticks(rotation=45)
plt.show()
```

```python
top_rest = clean_df["rest_type"].value_counts().head(10)
plt.figure(figsize = (8,5))
top_rest.plot(kind ="bar")
plt.xlabel("Resturant Type")
plt.title("Top 10 Restaurant Types")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
online  = clean_df.groupby("online_order")["rate"].mean()
plt.figure(figsize =(5,5))
online.plot(kind = "bar")
plt.title("Average Rating by Online Order")
plt.ylabel("Average Rating")
plt.show()
```

```python
booking = clean_df.groupby("book_table")["rate"].mean()

plt.figure(figsize=(5,5))
booking.plot(kind="bar")
plt.title("Average Rating by Table Booking")
plt.ylabel("Average Rating")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(clean_df["approx_cost(for two people)"], bins=20)
plt.title("Cost Distribution")
plt.xlabel("Approx Cost for Two")
plt.ylabel("Restaurant Count")
plt.show()
```

```python
top_cuisine = clean_df["cuisines"].value_counts().head(10)

plt.figure(figsize=(10,5))
top_cuisine.plot(kind="bar")
plt.title("Top 10 Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Restaurant Count")
plt.xticks(rotation=60)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(clean_df["votes"], clean_df["rate"])
plt.title("Votes vs Rating")
plt.xlabel("Votes")
plt.ylabel("Rating")
plt.show()
```
